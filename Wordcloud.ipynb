# Word Cloud

 A technique to select several words that occur frequently in a sentence and illustrate them in a size that corresponds to their frequency. \
  Refers to the automatic arrangement of words that appear frequently on a web page or blog, for example. \
 A method to give an impression of the content of a text at a glance by varying not only the size of the text but also its color, font style, and orientation.

# <span style="font-size: 100%; color: red;">English ver</span>

```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from PIL import Image
```

```python
#Eliminate unnecessary words
stop_words_en = ['et', 'is', 'of', 'and', 'the', 'to', 'it', 'al', 'a1', 'a'\
                 , 'Fig','for', 'in', 'as', 'or', 'are', 'be', 'this', 'that', 'will'\
                 , 'there', 'was', 'by', 'BF', 'at', 'with', 'et']

path = './English.txt'
with open(path, mode='r') as f:
    text = f.read()

wc1 = WordCloud(background_color="white", width=1000, height=600, stopwords=set(stop_words_en))

wc1.generate(text)
#plt.imshow(wc1)
wc1.to_file('word.png')
#img = wc1.to_array()
#print(img.shape) 
plt.show()
```

### Change of shape (circular)

```python
x, y = np.ogrid[:300, :300]

mask = (x - 150) ** 2 + (y - 150) ** 2 > 150 ** 2
mask = 255 * mask.astype(int)


wc2 = WordCloud(background_color="white", repeat=True, mask=mask, stopwords=set(stop_words_en))
wc2.generate(text)

plt.axis("off")
plt.imshow(wc2, interpolation="bilinear")
plt.savefig('word_Circle_1.png', dpi=1000)
wc2.to_file('word_Circle.png')
img = wc2.to_array()
print(img.shape) 
plt.show()
```

### Change of shape (speech bubble)

```python
mask = np.array(Image.open('bubble.png'))
#print(mask)
#mask = np.where(mask == 0, 0,255)
#print(mask)
```

```python
wc3 = WordCloud(mask = mask,
                      stopwords = set(stop_words_en),
                      font_path="/system/Library/Fonts/ W4.ttc",
                      colormap = 'cool',
                      width=800, height=600).generate(text)

wc3.generate(text)
wc3.to_file('word_bubble.png')
plt.imshow(wc3)
plt.show()
```
